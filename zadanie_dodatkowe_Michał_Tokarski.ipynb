```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
```

w przypadku rozk≈Çadu normalnego $a = 1$, $b_+ = \sqrt{\dfrac{2}{e}}$

```python
def normalny_ilorazu(n):
    normal = []
    for i in range(n):
        check = True
        while check:
            u1 = np.random.random()
            u2 = np.random.uniform(-np.sqrt(2/np.e),np.sqrt(2/np.e))
            X = u2/u1
            if u1**2 <= np.e**(-X**2/2):
                normal.append(X)
                check = False
    return normal
```

```python
n =100000
normal = normalny_ilorazu(n)
len(normal)
```

```python
plt.figure(figsize=(14,8))
plt.hist(normal, bins = 100, density=True, label = "gƒôsto≈õƒá empiryczna")
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
plt.plot(x,norm.pdf(x), label = "gƒôsto≈õƒá teoretyczna")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```

```python
normal = np.array(normal)
cdfx = np.sort(normal)
x_val = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = normal.size
y_val = []
for i in x_val:
    temp = normal[normal <= i]
    value = temp.size / size
    y_val.append(value)
```

```python
plt.figure(figsize=(14,8))
plt.plot(x_val,np.array(y_val), label = "empirical distribution")
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
plt.plot(x, norm.cdf(x), label = 'theoretical distribution')
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend()
```

Drugim rozk≈Çadem bƒôdzie rozk≈Çad wyk≈Çadniczy
w przypadku rozk≈Çadu wyk≈Çadniczego $a = 1$, $b_+ = \sqrt{\dfrac{2}{\lambda\mathrm{e}}}$

```python
def expon_ilorazu(n,lam):
    exponential = []
    for i in range(n):
        check = True
        while check:
            u1 = np.random.random()
            u2 = np.random.uniform(0,np.sqrt(2/(lam*np.e)))
            X = u2/u1
            if u1**2 <= np.e**(-X*lam):
                exponential.append(X)
                check = False
    return exponential
```

```python
n =100000
exponential = expon_ilorazu(n,1)
```

```python
plt.figure(figsize=(14,8))
plt.hist(exponential, bins = 100, density=True, label = "gƒôsto≈õƒá empiryczna")
x = np.linspace(expon.ppf(0.01),expon.ppf(0.99), 100)
plt.plot(x,expon.pdf(x), label = "gƒôsto≈õƒá teoretyczna")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```

```python
exponential = np.array(exponential)
cdfx = np.sort(exponential)
x_val = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = exponential.size
y_val = []
for i in x_val:
    temp = exponential[exponential <= i]
    value = temp.size / size
    y_val.append(value)

plt.figure(figsize=(14,8))
plt.plot(x_val,np.array(y_val), label = "empirical distribution")
x = np.linspace(expon.ppf(0.01),
                expon.ppf(0.99), 100)
plt.plot(x, expon.cdf(x), label = 'theoretical distribution')
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend()
```
