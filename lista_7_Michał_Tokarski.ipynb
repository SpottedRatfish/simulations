```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Poisson_niejednorodny(lam,T):
    t = 0
    S =[0]
    check = True
    while check:
        U = np.random.random()
        t = t - 1/lam*np.log(U)
        if t > T:
            check = False
        if np.random.random() <= t**2/lam:
            S.append(t)
    return S
```

```python
lam = 20**2
plt.figure(figsize=(14,8))
plt.xlim(0,20)
for i in range(100):
    x = Poisson_niejednorodny(lam,20)
    n = len(x)
    y = np.arange(n)
    plt.step(x,y-1,color = 'blue')
plt.plot(np.linspace(0,20,1000),1/3*np.linspace(0,20,1000)**3, color = 'red')
plt.legend(handles = [plt.Line2D([0], [0], color='b', lw=4, label='procesy Poissona'),plt.Line2D([0], [0], color='r', lw=4, label="1/3*t^3")])
```

```python
def Poisson_niejednorodny_2(T):
    t = 0
    S = [0]
    check = True
    while check:
        U = np.random.random()
        tau = ((1+t)/np.sqrt(-U+1)) - 1 - t
        t = t + tau
        if t > T:
            check = False
        S.append(t)
    return S
```

```python
plt.figure(figsize=(14,8))
plt.xlim(0,20)
for i in range(1000):
    x = Poisson_niejednorodny_2(20)
    n = len(x)
    y = np.arange(n)
    plt.step(x,y-1,color = 'blue')
plt.plot(np.linspace(0,20,1000),2*np.log(1+np.linspace(0,20,1000)), color = 'red')
plt.legend(handles = [plt.Line2D([0], [0], color='b', lw=4, label='procesy Poissona'),plt.Line2D([0], [0], color='r', lw=4, label="2ln(1+t)")])
```

```python
def Poisson_niejednorodny_3(T):
    n = np.random.poisson(1/2*T-1/4*np.sin(2*T))
    if n == 0:
        return 0
    V = []
    while len(V) < n:
        X = np.random.uniform(0,T)
        U = np.random.random()
        if U <= (np.sin(X)**2)*T:
            V.append(X)
    return np.sort(V)

Poisson_niejednorodny_3(np.pi*10)
#poprawić
```

```python
T = np.pi*10
plt.figure(figsize=(14,8))
plt.xlim(0,T)
for i in range(1000):
    x = Poisson_niejednorodny_3(T)
    n = len(x)
    y = np.arange(n)
    plt.step(x,y-1,color = 'blue')
plt.plot(np.linspace(0,T,1000),1/2*np.linspace(0,T,1000)-1/4*np.sin(2*np.linspace(0,20,1000)), color = 'red')
plt.legend(handles = [plt.Line2D([0], [0], color='b', lw=4, label='procesy Poissona'),plt.Line2D([0], [0], color='r', lw=4, label="1/2t-1/4sin(2t)")])
```
