```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate(n,wartości,prawd):
    x = np.random.random(n)
    prawd_różn = 1
    for i in range(len(prawd)):
        prawd_różn = prawd_różn - prawd[-i-1]
        x[(prawd_różn<x) & (x<prawd_różn + prawd[-i-1])] = wartości[-i-1] 
    return x
```

```python
prawd1 = [0.1,0.35,0.15,0.4]
wartości1 = [1,2,3,4]
N = 10000
X1 = simulate(N, wartości1, prawd1)
M1 = np.array([np.count_nonzero(X1 == x) for x in wartości1])/N
```

```python
def zadanie1(n):
    count = 0
    prawd4 = [0.15,0.2,0.45,0.2]
    prawd3 = [0.1,0.35,0.25,0.3]
    prawd1 = [0.1,0.35,0.15,0.4]
    prawd2 = [0.2,0.25,0.35,0.2]
    wartości = [1,2,3,4]
    for i in range(n):
        X = simulate(1,wartości,prawd1)
        if X == 1:
            X2 = simulate(1,wartości,prawd1)
            if X2 == 3:
                count += 1
        if X == 2:
            X2 = simulate(1,wartości,prawd2)
            if X2 == 3:
                count += 1
        if X == 3:
            X2 = simulate(1,wartości,prawd3)
            if X2 == 3:
                count += 1
        if X == 4:
            X2 = simulate(1,wartości,prawd4)
            if X2 == 3:
                count += 1
    return count/n
```

```python
def zadanie2(n):
    prawd4 = [0.15,0.2,0.45,0.2]
    prawd3 = [0.1,0.35,0.25,0.3]
    prawd1 = [0.1,0.35,0.15,0.4]
    prawd2 = [0.2,0.25,0.35,0.2]   
    prawdopodobieństwa = [prawd1,prawd2,prawd3,prawd4] 
    count = []
    for i in range(n):
        X = simulate(1,[1,2,3,4],prawd4)
        count2 = 1
        while X != 2:
            X = simulate(1,[1,2,3,4],prawdopodobieństwa[int(X)-1])
            count2 += 1
        count.append(count2)
    return np.mean(count)
```

```python
def bladzenie_losowe_jeden_krok(p_pocz,p):
    if np.random.random() <= p:
        p_pocz += 1
    else:
        p_pocz -= 1
    return p_pocz
```

```python
p = 1
def bladzenie_losowe_calosc(n,p,p_pocz):
    Sn = []    
    for i in range(n):
        p_pocz = bladzenie_losowe_jeden_krok(p_pocz,p)
        Sn.append(p_pocz)
    return Sn
```
