```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
n = 100000
U1 = np.random.random(n)
U2 = np.random.random(n)
X = np.sqrt(-2*np.log(U1))*np.cos(2*np.pi*U2)
arrgs = np.linspace(-4,4,100)
```

```python
plt.hist(X, bins=50,density=True)
plt.plot(arrgs,ss.norm.pdf(arrgs))
```

```python
X.mean()
```

```python
X.var()
```

```python
ss.skew(X)
```

```python
ss.kurtosis(X)
```

```python
def polar_method(n):
    results1 = []
    results2 = []
    for i in range(n):
        while True:
            U1 = np.random.uniform(-1,1)
            U2 = np.random.uniform(-1,1)
            Rsquared = U1**2 + U2**2
            if Rsquared <= 1:
                results1.append(np.sqrt(-2*np.log(Rsquared)/Rsquared)*U1)
                results2.append(np.sqrt(-2*np.log(Rsquared)/Rsquared)*U2)
                break
    return results1,results2
```

```python
polar = np.array([polar_method(1000)])
```

```python
polar.mean()
```

```python
polar.var()
```

```python
ss.kurtosis(polar_method(10000)[0])
```

```python
ss.skew(polar_method(10000)[0])
```
