```python
def erlang_emp(n, lamb, size):
    X = []
    for i in range(size):
        U = np.random.uniform(0,1,n)
        X.append((-1/lamb)*np.log(np.prod(U)))
    return X
```

```python
n = 5
lamb = 1
erlang_empirical = erlang_emp(n,lamb,100000)
plt.figure(figsize=(14,8))
plt.hist(erlang_empirical, bins=50, density=True, label='empirical density')
x_arr=np.linspace(0, max(erlang_empirical), 100)
plt.plot(x_arr, erlang.pdf(x_arr, n, 0, 1), 'r', label='theoretical density')
plt.legend()
plt.title('Erlang distribution density')
```

```python
def laplace_emp(lam,size):
    
    u = np.random.random(size)
    n= len(u[u<=0.5])
    X1 = -np.log(1-np.random.random(n)) / lam
    X2 = np.log(np.random.random(size-n)) / lam
    X = np.append(X2, X1)
    return X

lam=1
laplace_empirical = laplace_emp(lam, 10000)            
plt.figure(figsize=(14,8))
plt.hist(laplace_empirical, bins=100, density=True, label='empirical density')
x_arr=np.linspace(-max(laplace_empirical), max(laplace_empirical), 100)
plt.plot(x_arr, laplace.pdf(x_arr,0,1), 'r', label='theoretical density')
plt.legend()
plt.title('Laplace distribution density')
```

```python
#teraz druga całka:
n = 10000
def normal_accept_method(n):
    results = []
    for i in range(n):
        while True:
            X = np.random.exponential()
            Y = np.random.random()
            if Y <= (1/(np.sqrt(2*np.pi))*(np.exp((-X**2)/2+X)))*(np.sqrt(2*np.e/np.pi)):
                results.append(X)
                break
    return results

x_arr = 1/np.sqrt(2)*np.array([normal_accept_method(n)])
y_arr = 1/np.sqrt(2)*np.array([normal_accept_method(n)])
```
