```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
#zadanie 1 - gęstość f(x) = 3/2(1-x**2) na x od 0 do 1
```

```python
def f(n):
    results = []
    while len(results) < n:
        X = np.random.uniform(0,3/2)
        Y = np.random.random()
        if Y <= (1-X**2):
            results.append(Y)
    return results
```

```python
fx = f(10000)
```

```python
plt.hist(fx, bins=25, density=True, label = "empirical density")
x = np.linspace(0,1,1000)
plt.plot(x,3/2*(1-x**2), label = "theoretical density")
plt.legend()
```

```python
def secondf(n):
    results = []
    while len(results) < n:
        X = np.random.random()
        Y = np.pi*(np.random.random())**(1/3)
        if X <= np.sin(Y)**2/(Y**2):
            results.append(Y)
    return results
```

```python
second = secondf(10000)
```

```python
plt.hist(second,bins=25,density=True, label="empirical density")
x = np.linspace(0,np.pi,1000)
plt.plot(x,2/np.pi*(np.sin(x)**2), label = "theoretical density")
plt.legend()
```

```python
def f_counts(n):
    results = []
    counts = []
    count = 0
    while len(results) < n:
        X = np.random.random()
        Y = np.pi*(np.random.random())**(1/3)
        count += 1
        if X <= np.sin(Y)**2/(Y**2):
            results.append(Y)
            counts.append(count)
            count = 0
    return counts
```

```python
n=10000
iterations = f_counts(n)
M =max(iterations)
y = []
for i in range(1,M+1):
    y.append(scipy.stats.geom.pmf(i,3/(2*((np.pi)**2))))
```

```python
probs = []
for i in range(1,M+1):
    probs.append(iterations.count(i))
probs = np.array(probs)
probs = probs/n
```

```python
plt.scatter(np.arange(1,M+1),y, label = 'theoretical values')
plt.scatter(np.arange(1,M+1),probs, label = "empirical values")
plt.legend()
```
